```python
 import matplotlib.pyplot as pt
```

```python
x=[1,3,5,7,9]
y=[2,4,6,8,10]

pt.plot(x,y,marker='o',color='green',linewidth=2,linestyle='dashed')
pt.xlabel('X')
pt.ylabel('Y')
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=['jan','feb','mar','apr','may','june']
y=[200,400,450,600,650,900]
pt.plot(x,y,marker='*',color='green',linewidth=2,linestyle='dashed')
pt.title('months')
pt.xlabel('months')
pt.ylabel("amnt")
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=['jan','feb','mar','apr','may','june']
y=[200,400,450,600,650,900]
pt.bar(x,y,color='skyblue')
pt.title('months')
pt.xlabel('months')
pt.ylabel("amnt")
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=['jan','feb','mar','apr','may','june']
y=[200,400,450,600,650,900]
pt.bar(x,y,color=['skyblue','red','green'])
pt.title('months')
pt.xlabel('months')
pt.ylabel("amnt")
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=['jan','feb','mar','apr','may','june']
y=[200,400,450,600,650,900]
pt.bar(x,y,color=['skyblue','red','green'],edgecolor="purple")
pt.title('months')
pt.xlabel('months')
pt.ylabel("amnt")
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=[11,22,33,44,55,66,77,85,88,93,99]
y=[10,20,30,40]
pt.hist(x,bins=y,color='skyblue',edgecolor="purple")
pt.title('months')
pt.xlabel('months')
pt.ylabel("amnt")
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=[11,22,33,44,55,66,77,85,88,93,99]
y=[10,20,30,40]
pt.hist(x,bins=y,color='skyblue',edgecolor="purple")
pt.title('months')
pt.xlabel('months')
pt.ylabel("amnt")
pt.grid(True)
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=['jan','feb','mar','apr','may','june']
y=[20,40,50,60,70,90]
colors=['skyblue','red','green','brown','purple','orange']
pt.pie(y,labels=x,colors=colors,autopct='%1.1f%%')
pt.legend()
pt.title('piechart')
pt.show()
```

```python
 import matplotlib.pyplot as pt
x=[11,22,33,44]
y=[10,20,30,40]
pt.scatter(x,y,color="blue",s=100)
pt.title('scatter')
pt.show()
```
